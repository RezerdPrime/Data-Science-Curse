```python
q25 = df_nodub.year.quantile(0.25)
q75 = df_nodub.year.quantile(0.75)
iqr = q75 - q25

boards = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)

boards
```

```python
m = df_nodub.year[(df_nodub.year < boards[0]) | (df_nodub.year > boards[1])]
max(m.values)
```

```python
df_nodub.loc[(df_nodub.year < boards[0]), "year"] = int(boards[0])
df_nodub
```
